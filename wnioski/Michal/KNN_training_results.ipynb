# Plot results of randomized search on KNN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from time import time
from scipy.stats import randint as sp_randint
from sklearn import model_selection

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier

import pickle
import itertools
```

## Load pickled rsearch results

```python
KNN_10 = pickle.load( open( "KNN_10.pkl", "rb" ) )
KNN_100 = pickle.load( open( "KNN_100.pkl", "rb" ) )
rad_KNN_10 = pickle.load( open( "rad_NN_10.pkl", "rb" ) )
rad_KNN_100 = pickle.load( open( "rad_KNN_100.pkl", "rb" ) )
rad_NN_100_small = pickle.load( open ("rad_KNN_100.pkl", "rb"))
mah_KNN_100 = pickle.load( open( "mah_KNN_100.pkl", "rb" ) )
```

```python
print str(KNN_10.best_estimator_)
print str(KNN_100.best_estimator_)
print str(rad_KNN_10.best_estimator_)
print str(rad_KNN_100.best_estimator_)
```

## Results for KNN - 100 iterations

```python
results = KNN_100.cv_results_

feature = "n_neighbors"

plt.figure(figsize=(10, 5))
plt.title("RandomizedSearchCV for feature: %s" % feature, fontsize=16)

plt.xlabel(feature)
plt.ylabel("Score")
plt.grid()

ax = plt.axes()

# Get the regular numpy array from the |MaskedArray
X_axis = np.array(results['param_n_neighbors'].data, dtype=float)

sample_score_mean = results['mean_test_score']
sample_score_std = results['std_test_score']

lists = sorted(itertools.izip(*[X_axis, sample_score_mean, sample_score_std]))
new_x, new_y, new_std = list(itertools.izip(*lists))

new_x = np.array(list(new_x))
new_y = np.array(list(new_y))
new_std = np.array(list(new_std))

ax.fill_between(new_x, new_y - new_std,
                new_y + new_std,
                alpha=0.1)

ax.plot(new_x, new_y, alpha=1)

plt.show()
```

## KNN 100 - mean metric performance

```python
results = KNN_100.cv_results_

feature = "metric"

plt.figure(figsize=(10, 5))
plt.title("RandomizedSearchCV for feature: %s" % feature, fontsize=16)

plt.xlabel(feature)
plt.ylabel("Score")
plt.grid()

ax = plt.axes()

# Get the regular numpy array from the MaskedArray
X_axis = np.array(results['param_metric'].data, dtype=str)

sample_score_mean = results['mean_test_score']
sample_score_std = results['std_test_score']

pd_list = pd.DataFrame(
    {'metric': X_axis,
     'score': sample_score_mean,
     'std': sample_score_std
    })

print pd_list.groupby(pd_list.metric).mean()
```
