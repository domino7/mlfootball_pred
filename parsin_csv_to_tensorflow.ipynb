```python
import pandas as pd
import sklearn.model_selection as skl
import numpy as np

ML_DATASET = "learning_vectors01.csv"

data = pd.read_csv(ML_DATASET)

list(data.columns.values)
```

```python

names = ['H_Speed','H_Pass', 'H_Shoot', 'H_Pressure', 'H_chPass', 'H_chCross', 'H_dAggr', 'H_dWidth', 'A_Speed', 'A_Pass', 'A_Shoot', 'A_Pressure', 'A_chPass', 'A_chCross', 'A_dAggr', 'A_dWidth', 'H_age', 'A_age']

for name in names:
    data[name] = data[name] / 10
```

```python
data.head()
```

```python
names2 = ['H_TMV','A_TMV']

for name in names2:
    data[name] = data[name] / 100
    
data.iloc[2]
```

```python
names3 = ['League_id']

for name in names3:
    data[name] = data[name] / 1000
    
data.head()
```

```python
data.to_csv("learning_vectors03.csv")
```
